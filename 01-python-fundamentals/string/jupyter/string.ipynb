# 字符串

## 文本对齐

```python
'Hellow'.rjust(10,'-')
```

```python
'Hellow'.ljust(10,'-')
```

```python
'Hellow'.center(10,'-')
```

## 删除空白字符

```python
'  Hellow  '.lstrip()
```

```python
'  Hellow  '.rstrip()
```

```python
'  Hellow  '.strip()
```

```python
spam='spamSpamfssffffPamSsssSssspam'
spam.strip('spam')
```
