### 使用异常捕获处理

```python
try:
    print(alien_0['points'])
except KeyError:
    print("你要查询的键不存在!")
```

### 使用get()方法来获取

```python
points=alien_0.get('points')
points1=alien_0.get('points','并没有这个键')
x_position=alien_0.get('x_position','并没有这个键')
print(points)
print(points1)
print(x_position)
```

## 遍历字典

### items()同时遍历键值对

```python
my_friend={'first_name':'Chen','last_name':'YangFei','age':19,'city':'China'}
print("------------------------")
for key,value in my_friend.items():
    print(f"{key}:{value}")
print("------------------------")
```

### 仅仅遍历键或值

```python
#遍历字典时会默认遍历所有的键
for key in my_friend:
    print(f"{key}")
```

```python
for key in my_friend.keys():
    print(f"{key}")
```

```python
for value in my_friend.values():
    print(f"{value}")
```

## 字典的嵌套

### 列表中存储字典

```python
#处理很多外星人
aliens=[]
for alien_number in range(50):
    if alien_number%2==0:
        new_alien={'color':'green','speed':'slow','point':15}
        aliens.append(new_alien)
    else:
        new_alien={'color':'yellow','speed':'fast','point':40}
        aliens.append(new_alien)
```

```python
for alien in aliens[:6]:
    print(alien)
print("……")
print(f"Total number of aliens:{len(aliens)}")
```

```python
for alien in aliens[:6]:
    if alien['color']=='green':
        alien['color']='red'
        alien['point']=20
        alien['speed']='medium'
    print(alien)
print("……")
print(f"Total number of aliens:{len(aliens)}")
```

### 字典中嵌套字典

```python
users={
    'aeinstein':{
        'first_name':'albert',
        'last_name':'einstein',
        'location':'princetion',
    },
    'mcurie':{
        'first_name':'marie',
        'last_name':'curie',
        'location':'paris',
    },
}
print(users)
```

```python
for username,userinfo in users.items():
    print("-----------------------------------------")
    print(f"Username:{username}")
    print(f"\tFullname:{userinfo['first_name'].title()} {userinfo['last_name'].title()}")
    print(f"\tLocation:{userinfo['location'].title()}")
print("-----------------------------------------")
```

## setdefault()方法

```python
spam={'name':'Jon','age':20}
print(spam)
print("-----------------")
if 'color' not in spam:
    spam['color']='green'
print(spam)
```

```python
#效果等同于上面if语句
spam.setdefault('city','china')
print(spam)
```

```python
message="It was a bright cold day"
count={}
for character in message:
    count.setdefault(character,0)
    #如果已经存在并不会做什么修改如果不存在则添加
    count[character]=count[character]+1
print(count)
```
