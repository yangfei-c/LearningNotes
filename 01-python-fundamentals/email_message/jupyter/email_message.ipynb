# 发送电子邮件和短信

## SMTP电子邮件

```python
import smtplib
```

```python
smtObj=smtplib.SMTP('smtp.qq.com',587)
```

```python
type(smtObj)
```

```python
smtObj.ehlo()#向SMTP发送消息返回250则建立成功
```

```python
smtObj.starttls()#开始STL加密，返回220表示服务器已准备就绪
```

```python
smtObj.login('3260693417@qq.com','c15936982275')
```
