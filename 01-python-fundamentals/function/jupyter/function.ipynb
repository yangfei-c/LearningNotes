# 函数

## 在函数中修改列表

```python
def print_models(unprinted_designs,completed_models) ->None:
    while unprinted_designs:
        current_design=unprinted_designs.pop()
        print("--------------------------------")
        print(f"Printing model:{current_design}")
        completed_models.append(current_design)
```

```python
def show_models(models)->None:
    if models:
        for model in models:
            print("--------------------------------")
            print(model)
    else:
        print("--------------------------------")
        print("列表为空！")
```

```python
unprinted_designs=['phone case','robot pendant','dodecahedron']
completed_models=[]
print_models(unprinted_designs,completed_models)
show_models(completed_models)
show_models(unprinted_designs)
```

## 传递任意数量的实参

```python
def mak_pizza(*toppings)->None:#制作披萨并不知道传多少调料，
    # *作用，创建一个名为toppings的空元组
    print(toppings)
mak_pizza('pepperoni')
mak_pizza('pepperoni','mushroom','extra cheese')
```

## 使用任意数量的关键字实参

```python
#有时传递任意数量的实参但预先不知道传递给函数什么样的信息
#这时可以将函数编写出能够接受任意数量的键值对，
# **user_info会让python创建一个名为user_info的空字典
def build_profile(first_name,last_name,**user_info)->dict:
    user_info['first_name']=first_name
    user_info['last_name']=last_name
    return user_info
user_profile=build_profile('albert','einstein',location='china',age=20,sex='man')
print(user_profile)
```
