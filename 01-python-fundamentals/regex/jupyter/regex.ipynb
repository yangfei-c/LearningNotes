# 正则表达式

## 匹配电话号码

```python
import re
```

```python
phoneNumber_regex=re.compile(r'\d{3}-\d{3}-\d{4}')
prompt="please input your phone number" \
       "(enter quit to stop program):"
flag=True
while flag:
    phoneNumber=input(prompt)
    mo=phoneNumber_regex.search(phoneNumber)
    if mo:
        print(f"your number is {mo.group()}")
    else:
        if phoneNumber=='quit':
            flag=False
        else:
            flag=True
            print("你输入的不是一个正确电话号码")
```

### 添加括号将创建分组

```python
phoneNumber_regex=re.compile(r'(\d{3})-(\d{3}-\d{4})')
```

```python
mo=phoneNumber_regex.search("niaho 333-242-3321jfjjfjfjfjfe")
print(f"your number_city is {mo.group(1)} and number is {mo.group(2)}")
```

```python
mo.groups()
```

```python
mo.group(0)
```

### 使用\进行转义

```python
phoneNumber_regex=re.compile(r'(\(\d{3}\)) (\d{3}-\d{4})')
mo=phoneNumber_regex.search("niaho (333) 242-3321 jfjjfjfjfjfe")
```

```python
print(mo.groups())
```

```python
mo.group(1)
```

```python
mo.group(2)
```

## 使用管道匹配多个分组

```python
text_regex=re.compile(r'Bast|min')
mo=text_regex.search("Iam Min Bast and min love bast")
mo
```

```python
text_regex1=re.compile(r'Bat(mob|min|luc)')
mo=text_regex1.search("Iam batmob and Bat min love Batluc and Batmob and Batmob")
mo.group(1)
```

## 使用 ？实现可选匹配

```python
select_regex=re.compile(r'(\d-)bat(me)?man')
mo=select_regex.search("batman hellow")
mo
```

```python
mo=select_regex.search("1-batmman 1-batmeman hellow")
mo.group()
```

## * 实现零次或多次匹配

```python
many_regex=re.compile(r'bat(ni)*man')
mo=many_regex.search(" Batniman batnimen batninininiman")
mo
```

## +号实现匹配一次或多次

```python
must_regex=re.compile(r'bat(mi)+wed')
mo=must_regex.search("batwed batmiwed")
mo
```

```python
mo=must_regex.search("batwed")
mo==None
```

## 花括号匹配特定次数

```python
specific_regex=re.compile(r'hene(ni){3,7}')
mo=specific_regex.search("fjdjjheneninijdjj")
mo==None
```

```python
specific_regex=re.compile(r'hene(ni){3,7}')
mo=specific_regex.search("fjdjjheneninininininijdjj")
mo.group()
```

## 贪心匹配非贪心匹配
为什么上面返回henenininininini而不返回heneninini这也满足了最少三个最多七个的范围之内
这就是python正则表达式的贪心匹配
要想返回最少的即heneninini可以加个？，以变成非贪心匹配

```python
nongreedy_regex=re.compile("he(ni){3,7}?")
mo=nongreedy_regex.search("henininininini")
mo
```

## Regex对象的findall方法
search方法只能找到第一次匹配到的，对于文本中重复符合的可以利用findall来返回所有符合匹配的

```python
all_regex=re.compile(r'(he)(mn){3,5}')
all_regex1=re.compile(r'hemn{1,5}')
mo=all_regex.findall("hemnmnmnhemnmnmn")
mo1=all_regex1.findall("hemnhemnhemn")
print(mo)
print(mo1)
```

## 建立自己的字符分类

```python
mine_regex=re.compile(r'[abcd]')
mo=mine_regex.findall("jfjfiwefnviwfowefdsfjgbnvncmqazedv")
mo
```

```python
mine_regex1=re.compile(r'[^a-m]')
mo=mine_regex1.findall("jfjfiwefnviwfowefdsfjgbnvncmqazedv")
mo
```

## 匹配以特定字符开始或特定结尾的字符串

```python
begin_endwith_regex=re.compile(r'^\d$')
mo=begin_endwith_regex.search("3")
mo
```

```python
mo=begin_endwith_regex.search("3242")
mo==None
```

```python
begin_endwith_regex=re.compile(r'^\d+$')
mo=begin_endwith_regex.search("3435")
mo
```

## 不区分大小写的匹配

```python
non_discrimative_regex=re.compile(r'hellow',re.I)
mo=non_discrimative_regex.findall("HEllow,Hellow,hellow")
mo
```
