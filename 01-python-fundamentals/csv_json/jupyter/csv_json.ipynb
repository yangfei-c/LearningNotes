# json操作

```python
import json
```

## loads---load string 读取为python值

```python
json_data='{"name":"zop","iscat":"true","mice":0,"filea":"null"}'
json_data_as_python_value=json.loads(json_data)
json_data_as_python_value
```

## dumps---dump string读取为json值

```python
python_value={'name': 'zop', 'iscat': 'true', 'mice': 0, 'filea': 'null'}
json_value=json.dumps(python_value)
json_value
```
